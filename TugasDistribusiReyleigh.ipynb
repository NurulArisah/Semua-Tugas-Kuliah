```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# jumlah sample
n = 10000

# parameter scale
sigma = 1.0

# generate random number distribusi Rayleigh
data = np.random.rayleigh(sigma, n)
```

```python
# Scatterplot

plt.scatter(range(n), data, alpha=0.5, s=5)
plt.title("Scatterplot Distribusi Rayleigh")
plt.xlabel("Percobaan ke-")
plt.ylabel("Nilai Distribusi")
plt.show()
```

```python
# Histogram

plt.hist(data, bins=100, edgecolor='black', alpha=0.7)
plt.title("Histogram Distribusi Rayleigh")
plt.xlabel("Nilai Distribusi")
plt.ylabel("Frekuensi")
plt.show()
```
